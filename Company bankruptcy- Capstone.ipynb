# Function to remove features which are highly correlated

```python
def correlation_check(data,val):
    features=set()
    corr_matrix=data.corr()
    for i in range(len(corr_matrix)):
        for j in range(i):
            if corr_matrix.iloc[i,j]>val:
                features.add(corr_matrix.columns[i])
    return features
```

```python
features=correlation_check(data,0.8)
```

this code defines a function to identify pairs of features in a dataset that have a correlation coefficient above a specified threshold. It then returns a set of feature names that are highly correlated according to the given threshold. This can be useful for feature selection or further analysis to avoid multicollinearity issues in predictive modeling.

```python
features
```

```python
data.drop(features,axis=1,inplace=True)
```

# SMOTE : OverSampling

SMOTE (Synthetic Minority Over-sampling Technique) is a technique used in machine learning and data preprocessing to address class imbalance in a dataset. Class imbalance occurs when one class (usually the minority class) has significantly fewer examples than another class (majority class). This imbalance can lead to biased model performance and reduced predictive accuracy for the minority class.


Balance the data By using SMOTE

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
 y_resampled.value_counts()
```

```python
sns.countplot(x=y_resampled)
plt.show()
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_resampled)
```

# Principal Component Analysis

Principal Component Analysis (PCA) is a dimensionality reduction technique used in data analysis and machine learning to transform high-dimensional data into a lower-dimensional space while preserving as much of the data's variability as possible. It achieves this by identifying the principal components, which are orthogonal (uncorrelated) linear combinations of the original features. These principal components are ordered in such a way that the first component captures the maximum variance in the data, the second component captures the second maximum variance, and so on.

Perform PCA for dimensionality reduction

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=25) 
X_pca = pca.fit_transform(X_scaled)
```

The code applies Principal Component Analysis (PCA) with 25 components to the scaled dataset X_scaled. It transforms the data into a new representation called X_pca, which retains the most important information while reducing the original data's dimensionality.

```python
X_pca.shape
```

# Train-Test Split

# Logistic Regression

Logistic Regression is a fundamental and widely used statistical method in machine learning, primarily employed for binary classification tasks (classifying instances into one of two classes). Despite its name, logistic regression is used for classification rather than regression.

```python
from sklearn.linear_model import LogisticRegression
#create a model
classifier=LogisticRegression()
#fitting training data into the model
classifier.fit(X_train, y_train)
y_pred=classifier.predict(X_test)
print(y_pred)
print(list(zip(y_test,y_pred)))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
 
cfm=confusion_matrix(y_test,y_pred)
print(cfm)
 
print('classification report: ',)
print(classification_report(y_test,y_pred))
 
acc=accuracy_score(y_test,y_pred)
print('accuracy of the model: ',acc)
```

Precision-----> column Recall----->rows F-1----------> harmonic mean of precision and recall

# ADJUSTING THE THRESHHOLD

In logistic regression, the output is a probability value between 0 and 1 that represents the likelihood of a sample belonging to a particular class. To make predictions, you need to convert these probabilities into actual class labels. The default threshold for most binary classification algorithms is 0.5, meaning that if the predicted probability is greater than or equal to 0.5, the sample is classified as the positive class; otherwise, it's classified as the negative class.

However, adjusting the threshold can be important in cases where you want to optimize for specific metrics like precision, recall, or F1-score.

```python
acc=accuracy_score(y_test,y_pred)
print('accuracy of the model: ',acc)
```

```python
#store the predicted probabilities
y_pred_prob=classifier.predict_proba(X_test)
print(y_pred_prob)
```

```python
for a in np.arange(0.4,0.61,0.01):
    predict_mine = np.where(y_pred_prob[:,1] > a, 1, 0)
    cfm=confusion_matrix(y_test, predict_mine)
    total_err=cfm[0,1]+cfm[1,0]
    print("Errors at threshold ", a, ":",total_err, " , type 2 error :", 
          cfm[1,0]," , type 1 error:", cfm[0,1])
```

On 0.45 threshold overall error is 304 also type 2 error is also less than actual prediction.

```python
y_pred_class=[]
for value in y_pred_prob[:,1]:
    if value > 0.45:
        y_pred_class.append(1)
    else:
        y_pred_class.append(0)
print(y_pred_class)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
 
cfm=confusion_matrix(y_test,y_pred_class)
print(cfm)
 
print('classification report: ',)
print(classification_report(y_test,y_pred_class))
 
acc=accuracy_score(y_test,y_pred_class)
print('accuracy of the model: ',acc)
```

# Decision Tree Classifier

A Decision Tree is a widely used machine learning algorithm for both classification and regression tasks. It's a tree-like structure where each internal node represents a feature or attribute, each branch represents a decision rule, and each leaf node represents an outcome. Decision Trees are known for their ability to handle both categorical and numerical data and for their interpretability.

```python
from sklearn.tree import DecisionTreeClassifier

model_DecisionTree=DecisionTreeClassifier(random_state=10)

#fit the model on data and predict the values

model_DecisionTree.fit(X_train,y_train)

y_pred=model_DecisionTree.predict(X_test)

print(y_pred)
print(list(zip(y_test,y_pred)))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(y_test,y_pred)
print(cfm)

print('classification report: ',)
print(classification_report(y_test,y_pred))

acc=accuracy_score(y_test,y_pred)
print('accuracy of the model: ',acc)
```

Hyperparameters for tuning

```python
# Define hyperparameters for tuning
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

These hyperparameters collectively control the complexity, depth, and structure of the decision tree or random forest classifier. By exploring various combinations of these hyperparameter values, you can find the configuration that results in the best performance for your specific dataset.

```python
# Perform GridSearchCV for hyperparameter tuning
grid_search = GridSearchCV(model_DecisionTree, param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
# Get the best parameters
best_params = grid_search.best_params_
```
